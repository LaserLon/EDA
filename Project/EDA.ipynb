```python
# df_greenhouse_gas.reset_index(inplace=True)
# df_greenhouse_gas
```

```python
df_greenhouse_gas = df_greenhouse_gas.rename(columns={'year': 'Year'})
df_greenhouse_gas = df_greenhouse_gas.rename(columns={'value': 'GHG Kilotons'})
df_greenhouse_gas
```

```python
# df_greenhouse_gas.info()
# df_tempchange.info()

df_tempchange['Year'] = df_tempchange['Year'].astype(int)
df_data = df_tempchange.merge(df_greenhouse_gas, on=['Country', 'Year'], how='inner')
df_data = df_data.dropna().reset_index(drop=True)
df_data
```

```python
# df_data.set_index(['Country', 'Year'], inplace=True)
df_data.plot( figsize=(25,6))
```

```python
df_country_continents = pd.read_excel (r'./datasets/Contenent Country List.xlsx')
df_country_continents
```

```python

df_data = df_data.merge(df_country_continents, on='Country', how='left')
df_data.head()
```

```python
df_global = df_data.groupby('Year').agg({'Temperature Change': 'mean','GHG Kilotons': 'sum'}).reset_index()
df_global
```

```python
# Calculate the cumulative temperature change
df_global['Cumulative Temperature Change'] = df_global['Temperature Change'].cumsum()
df_global
```

```python

import matplotlib.pyplot as plt

fig, ax1 = plt.subplots()

color = 'tab:blue'
ax1.set_xlabel(df_global.columns[0])  # Year
ax1.set_ylabel(df_global.columns[1], color=color)  # Temperature Change
ax1.plot(df_global['Year'], df_global['Temperature Change'], color=color)
ax1.tick_params(axis='y', labelcolor=color)

ax2 = ax1.twinx()  # instantiate a second axes that shares the same x-axis

color = 'tab:red'
ax2.set_ylabel(df_global.columns[2], color=color)  # GHG Kilotons
ax2.plot(df_global['Year'], df_global['GHG Kilotons'], color=color)
ax2.tick_params(axis='y', labelcolor=color)

fig.tight_layout()  # otherwise the right y-label is slightly clipped
plt.title('Temperature Change and GHG Kilotons Over Years')
plt.show()
```

```python
df_global['Cumulative Temperature Change'] = df_global['Temperature Change'].cumsum()

# Plot the data
fig, ax1 = plt.subplots()

color = 'tab:blue'
ax1.set_xlabel(df_global.columns[0])  # Year
ax1.set_ylabel('Cumulative ' + df_global.columns[1], color=color)  # Cumulative Temperature Change
ax1.plot(df_global['Year'], df_global['Cumulative Temperature Change'], color=color)
ax1.tick_params(axis='y', labelcolor='tab:blue')

ax2 = ax1.twinx()  # instantiate a second axes that shares the same x-axis

color = 'tab:red'
ax2.set_ylabel(df_global.columns[2], color=color)  # GHG Kilotons
ax2.plot(df_global['Year'], df_global['GHG Kilotons'], color=color)
ax2.tick_params(axis='y', labelcolor=color)

fig.tight_layout()  # otherwise the right y-label is slightly clipped
plt.title('Global Temperature Change and GHG Kilotons Over Years')
plt.show()
```
