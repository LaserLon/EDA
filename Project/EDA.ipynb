```python
df_greenhouse_gas = df_greenhouse_gas[df_greenhouse_gas['category'] == 'greenhouse_gas_ghgs_emissions_without_land_use_land_use_change_and_forestry_lulucf_in_kilotonne_co2_equivalent']
df_greenhouse_gas
```

```python
df_greenhouse_gas = df_greenhouse_gas.drop(columns=['category'])
df_greenhouse_gas
```

```python
# Add a row for each year that is called world and contains a sum of all other contries in that year

df_world = df_greenhouse_gas.groupby('year')['value'].sum().reset_index()
df_world['Country'] = 'World'
df_world = df_world[['Country', 'year', 'value']]
df_combined = pd.concat([df_greenhouse_gas, df_world], ignore_index=True)
df_greenhouse_gas = df_combined.sort_values(by=['year', 'Country']).reset_index(drop=True)
df_greenhouse_gas
```

```python
df_greenhouse_gas = df_greenhouse_gas.rename(columns={'year': 'Year'})
df_greenhouse_gas = df_greenhouse_gas.rename(columns={'value': 'GHG Kilotons'})
df_greenhouse_gas
```

```python

import matplotlib.pyplot as plt
pd.options.display.float_format = '{:.2f}'.format

fig, ax1 = plt.subplots()

color = 'tab:blue'
ax1.set_xlabel(df_global.columns[1])  # Year
ax1.set_ylabel(df_global.columns[2], color=color)  # Temperature Change
ax1.plot(df_global['Year'], df_global['Temperature Change'], color=color)
ax1.tick_params(axis='y', labelcolor=color)
ax1.set_ylim(0, max(df_global['Temperature Change']) * 2)


ax2 = ax1.twinx()

color = 'tab:red'
ax2.set_ylabel(df_global.columns[3], color=color)  # GHG Kilotons
ax2.plot(df_global['Year'], df_global['GHG Kilotons'], color=color)
ax2.tick_params(axis='y', labelcolor=color)
ax2.set_ylim(0, max(df_global['GHG Kilotons']) * 1.1)  


fig.tight_layout()  # otherwise the right y-label is slightly clipped
plt.title(f'World\nGHG Kilotons vs Temperature Change Over Years')
plt.show()
```

```python

df_temp = df_data.loc[df_data['Continent'] != 'World']
df_continent = df_temp.groupby('Continent')['GHG Kilotons'].sum().reset_index()
df_pareto = df_continent.sort_values(by='GHG Kilotons', ascending=False)
df_pareto['Cumulative Percentage'] = df_pareto['GHG Kilotons'].cumsum() / df_pareto['GHG Kilotons'].sum() * 100

fig, ax1 = plt.subplots()
ax1.bar(df_pareto['Continent'], df_pareto['GHG Kilotons'], color='C0')
ax1.set_xlabel('Continent')
ax1.set_ylabel('GHG Kilotons', color='C0')
ax1.tick_params(axis='y', labelcolor='C0')

ax2 = ax1.twinx()
ax2.plot(df_pareto['Continent'], df_pareto['Cumulative Percentage'], color='C1', marker='D', ms=7)
ax2.set_ylabel('Cumulative Percentage', color='C1')
ax2.tick_params(axis='y', labelcolor='C1')
ax2.set_ylim(0, 100)

plt.title('Pareto Chart of GHG Kilotons by Continent')
fig.tight_layout()
plt.show()
```

```python
df_temp = df_data.loc[df_data['Continent'] != 'World']
df_continent_year = df_temp.groupby(['Continent', 'Year'])['GHG Kilotons'].sum().reset_index()
years = df_continent_year['Year'].unique()
num_years = len(years)

fig, axes = plt.subplots(num_years, 1, figsize=(10, 5*num_years), sharex=True)

# Iterate through each year
for i, year in enumerate(years):
    ax1 = axes[i]
    df_year = df_continent_year[df_continent_year['Year'] == year].sort_values(by='GHG Kilotons', ascending=False)
    df_year['Cumulative Percentage'] = df_year['GHG Kilotons'].cumsum() / df_year['GHG Kilotons'].sum() * 100

    ax1.bar(df_year['Continent'], df_year['GHG Kilotons'], color='C0')
    ax1.set_ylabel('GHG Kilotons', color='C0')
    ax1.tick_params(axis='y', labelcolor='C0')
    ax1.set_title(f'Pareto Chart of GHG Kilotons by Continent for {year}')
    
    ax2 = ax1.twinx()
    ax2.plot(df_year['Continent'], df_year['Cumulative Percentage'], color='C1', marker='D', ms=7)
    ax2.set_ylabel('Cumulative Percentage', color='C1')
    ax2.tick_params(axis='y', labelcolor='C1')
    ax2.set_ylim(0, 100)

axes[-1].set_xlabel('Continent')

plt.tight_layout()
plt.show()
```

```python
df_temp = df_data.loc[df_data['Country'] != 'World']
df_country = df_temp.groupby('Country')['GHG Kilotons'].sum().reset_index()
df_pareto = df_country.sort_values(by='GHG Kilotons', ascending=False).head(10)
df_pareto['Cumulative Percentage'] = df_pareto['GHG Kilotons'].cumsum() / df_pareto['GHG Kilotons'].sum() * 100

fig, ax1 = plt.subplots(figsize=(10, 6))

ax1.bar(df_pareto['Country'], df_pareto['GHG Kilotons'], color='C0')
ax1.set_xlabel('Country')
ax1.set_ylabel('GHG Kilotons', color='C0')
ax1.tick_params(axis='y', labelcolor='C0')

ax2 = ax1.twinx()
ax2.plot(df_pareto['Country'], df_pareto['Cumulative Percentage'], color='C1', marker='D', ms=7)
ax2.set_ylabel('Cumulative Percentage', color='C1')
ax2.tick_params(axis='y', labelcolor='C1')
ax2.set_ylim(0, 100)

plt.title('Pareto Chart of GHG Kilotons by Top 10 Countries')
fig.tight_layout()
plt.show()
```
